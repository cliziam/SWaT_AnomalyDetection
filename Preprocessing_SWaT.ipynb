# **SWaT Dataset**

**Secure Water Treatment (SWaT)** <br>

Il dataset Secure Water Treatment (SWaT) è un dataset multivariato di serie temporali che contiene dati raccolti da 51 sensori e attuatori in una centrale di trattamento dell'acqua. <br>
Il dataset è stato raccolto nel corso di un periodo di 11 giorni, con 7 giorni di funzionamento normale e 4 giorni di scenari di attacco. 
<br>
I dati sono etichettati in base a comportamenti normali e anomali, e gli scenari di attacco sono stati derivati attraverso modelli di attacco sviluppati dal team di ricerca. <br>
Sono stati lanciati 41 attacchi durante i 4 giorni di scenari di attacco.

Tutte le [informazioni sul dataset al link qui](https://itrust.sutd.edu.sg/itrust-labs_datasets/dataset_info/)<br>
Ho fatto richiesta al link citato nel paper *Deep Learning for Anomaly Detection in Time-Series Data: Review, Analysis, and Guidelines.*

### **1. Check missing values and drop_duplicates**

```python
def check(df):
    print('\n')
    if df.duplicated().any():
        df.drop_duplicates(inplace=True)
        print('Duplicates removed from', df.name)
    else:
        print('No duplicates found in', df.name)
    
    if df.isna().sum().any():
        print('Null values found in', df.name)
    else:
        print('No null values found in', df.name)
```
